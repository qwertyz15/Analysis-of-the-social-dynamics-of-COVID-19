# ***Beautiful Soap***


```
from bs4 import BeautifulSoup
import requests
```

```
source = requests.get('https://www.prothomalo.com/bangladesh/article/1656524/%E0%A7%A9%E0%A7%A6-%E0%A6%AE%E0%A7%87-%E0%A6%AA%E0%A6%B0%E0%A7%8D%E0%A6%AF%E0%A6%A8%E0%A7%8D%E0%A6%A4-%E0%A6%AF%E0%A6%BE%E0%A6%A4%E0%A7%8D%E0%A6%B0%E0%A7%80%E0%A6%AC%E0%A6%BE%E0%A6%B9%E0%A7%80-%E0%A6%AB%E0%A7%8D%E0%A6%B2%E0%A6%BE%E0%A6%87%E0%A6%9F-%E0%A6%9A%E0%A6%B2%E0%A6%BE%E0%A6%9A%E0%A6%B2-%E0%A6%AC%E0%A6%A8%E0%A7%8D%E0%A6%A7').text
soup = BeautifulSoup(source, 'lxml')
```

# **Title**

```
title = soup.title.text
title
```

#**Articles**


```
soup.find_all('p')[-1].extract()

article = soup.find_all('p')
for x in article:
    print(x.text)
```

# **Spans**

```
spans = soup.find_all('span')
for x in spans:
    print(x.text)
```
