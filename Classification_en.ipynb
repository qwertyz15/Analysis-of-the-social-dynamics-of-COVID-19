```
import pandas as pd
import numpy as np
df = pd.read_csv('data_clean_en.csv') #from 8 topic LDA
df
```

```
df.dropna(inplace = True)
df.reset_index(drop = True, inplace=True)
```

```
df
```

```
df['Dominant Topic'].unique()
```

```
corpus = df['data_ready'].copy()
```

```
# Creating the Bag of Words model
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=10000)
X = cv.fit_transform(corpus).toarray()
```

```
X.shape
```

```
y=df['Dominant Topic'].values
y
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```
X_train.shape,y_train.shape
```

# Decision Tree

```
## Classifier
from sklearn.tree import DecisionTreeClassifier
decision_Tree = DecisionTreeClassifier(criterion='entropy').fit(X_train, y_train)

y_pred = decision_Tree.predict(X_test)
```

```
from sklearn.metrics import confusion_matrix
cm1 = confusion_matrix(y_test, y_pred)
cm1
```

```
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score

print("Accuracy score: ", accuracy_score(y_test, y_pred))
print("Recall score: ", recall_score(y_test, y_pred, average = 'weighted'))
print("Precision score: ", precision_score(y_test, y_pred, average = 'weighted'))
print("F1 score: ", f1_score(y_test, y_pred, average = 'weighted'))

dt = []
dt.append(accuracy_score(y_test, y_pred))
dt.append(recall_score(y_test, y_pred, average = 'weighted'))
dt.append(precision_score(y_test, y_pred, average = 'weighted'))
dt.append(f1_score(y_test, y_pred, average = 'weighted'))
```
