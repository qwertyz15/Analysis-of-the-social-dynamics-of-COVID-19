```
df = df[:-4]
```

```
df['Date'] = pd.to_datetime(df['Date'], utc=False)
```

```
df = df.set_index('Date')
```

```
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")
fig, ax1 = plt.subplots(1,1, figsize=(21,12))
plt.plot(df.index, df['Daily News'], label='Daily News',color = 'green')
plt.bar(df.index, df['Daily Cases'], label='Daily Cases',color = 'blue')
plt.bar(df.index, df['Daily Death'], label='Daily Death',color = 'red')

plt.xlabel('Date')
plt.ylabel('Counts')
plt.title('Daily News & Cases (21 January 2020 - 19 May 2020)')
plt.xticks(rotation=90)

plt.legend()

plt.tight_layout()

plt.savefig('plot01.png')

plt.show()
```
