```
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")

slices1 = df1['Counts'].tolist()
labels1 = df1['Topic'].tolist()
#explode = [0, 0, 0, 0, 0]
colors1 = ['#e6194b', '#3cb44b', '#ffe119', '#4363d8', '#f58231', '#911eb4', '#46f0f0', '#f032e6']
plt.figure(figsize= (12,6))
plt.pie(slices1, 
        labels=labels1, 
        #explode=explode, 
        shadow=True,
        startangle=90, 
        autopct='%1.1f%%',
        wedgeprops={'edgecolor': 'black'},
        colors = colors1
        )

plt.title("Pie Chart : Topic")
plt.tight_layout()
plt.savefig('Topic01.png')
plt.show()
```
